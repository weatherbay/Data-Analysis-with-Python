```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importing data set
df = pd.read_csv('medical_examination.csv')
```

```python
df
```

Adding overweight column to the table by determining the BMI of each person

```python
Weight_per_person = df['weight']
Height_per_person = df['height']/100
df['BMI_per_person'] = round(Weight_per_person/(pow(Height_per_person,2)))

```

```python
My_list = []
for i in df['BMI_per_person']:
    if i > 25:
        My_list.append(1)
    else:
        My_list.append(0)
#print(My_list)
df['overweight'] = My_list        
    
```

```python
df
```

Noramlize data by making 0 as good and 1 as bad. implement for cholesterol and gluc

```python
My_list_1 = []
My_list_2 = []

for i in df['cholesterol']:
    if i == 1:
        My_list_1.append(0)
    else:
        My_list_1.append(1)
df['cholesterol'] = My_list_1

for i in df['gluc']:
    if i ==1:
        My_list_2.append(0)
    else:
        My_list_2.append(1)
df['gluc'] = My_list_2

New_df = df
```

```python
New_df

```

Drawing the categorical plot

```python
Cholesterol_good = []
Cholesterol_good_count = 0
Cholesterol_bad_count = 0
Cholesterol_bad = []

Gluc_good = []
Gluc_good_count = 0
Gluc_bad_count = 0
Gluc_bad = []

Active_good = []
Active_good_count = 0
Active_bad_count = 0
Active_bad = []

Cardio_good = []
Cardio_good_count = 0
Cardio_bad_count = 0
Cardio_bad = []

Smoke_good = []
Smoke_good_count = 0
Smoke_bad_count = 0
Smoke_bad = []

Alco_good = []
Alco_good_count = 0
Alco_bad_count = 0
Alco_bad = []

for i in New_df['cholesterol']:
    if i == 0:
        Cholesterol_good.append(i)
    else:
        Cholesterol_bad.append(i)

for i in New_df['gluc']:
    if i == 0:
        Gluc_good.append(i)
    else:
        Gluc_bad.append(i)

for i in New_df['active']:
    if i == 0:
        Active_good.append(i)
    else:
        Active_bad.append(i)

for i in New_df['cardio']:
    if i == 0:
        Cardio_good.append(i)
    else:
        Cardio_bad.append(i)

for i in New_df['smoke']:
    if i == 0:
        Smoke_good.append(i)
    else:
        Smoke_bad.append(i)
        
for i in New_df['alco']:
    if i == 0:
        Alco_good.append(i)
    else:
        Alco_bad.append(i)


Gluc_good_count = len(Gluc_good)
Gluc_bad_count = len(Gluc_bad)
Cholesterol_good_count = len(Cholesterol_good)
Cholesterol_bad_count = len(Cholesterol_bad)
Active_good_count = len(Active_good)
Active_bad_count = len(Active_bad)
Cardio_good_count = len(Cardio_good)
#print(Cardio_good_count)
Cardio_bad_count = len(Cardio_bad)
#print(Cardio_bad_count)
Smoke_good_count = len(Smoke_good)
#print(Smoke_good_count)
Smoke_bad_count = len(Smoke_bad)
#print(Smoke_bad_count)



    
```

```python

data = {'Category':['Gluc','Gluc','Cholesterol','Cholesterol','Active','Active','Cardio','Cardio'],
       'Subcategory':['Gluc=0','Gluc=1','Cholesterol=0','Cholesterol=1','Active=0','Active=1','Cardio=0','Cardio=1'],
       'Count':[Gluc_good_count,Gluc_bad_count,Cholesterol_good_count,Cholesterol_bad_count,Active_good_count,Active_bad_count,Cardio_good_count,Cardio_bad_count]}
Plotting_data = pd.DataFrame(data)

#creating a stacked bar plot
sns.set(style="whitegrid")
g = sns.catplot(data =Plotting_data, kind="bar",x="Category",y="Count",hue="Subcategory",ci="sd",palette="dark",alpha=.6,height=5)
g.despine(left=True)
g.set_axis_labels("Category","Count")
g.legend.set_title("Subcategories")
plt.figure(figsize=(20,5))
plt.show()
Fig_plot = g.savefig('catplot.png')
```
