```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importing data set
df = pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

Create a scatter plot

```python
sns.jointplot(x="Year", y="CSIRO Adjusted Sea Level", data=df, size=9)
```

```python
fig, ax = plt.subplots(figsize=(18, 7))
sns.regplot(x="Year", y="CSIRO Adjusted Sea Level", data=df, ax=ax)
```

```python
Adjusted_data = df[df['Year'] >= 2000]
```

```python
Adjusted_data
fig, ax = plt.subplots(figsize=(18, 7))
sns.regplot(x="Year", y="CSIRO Adjusted Sea Level", data=Adjusted_data,ax=ax)
plt.title("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("Sea level (inches)")
plt.savefig('sea_level_plot.png')
```
